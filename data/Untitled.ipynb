```python
lst_1 = [i for i in range(100)]
lst_2 = [i for i in range(100, 200)]
```

```python
%%time

for num1, num2 in zip(lst_1, lst_2):
    print(f'num1: {num1}, num2: {num2}')
```

```python
%%time

for i in range(len(lst_1)):
    print(f'num1: {lst_1[i]}, num2: {lst_2[i]}')
```

```python
%%time

for i, num1 in enumerate(lst_1):
    print(f'num1: {num1}, num2: {lst_2[i]}')
```

```python
from itertools import zip_longest

lst_3 = [i for i in range(10)]
lst_4 = [i for i in range(20)]
```

```python
for num3, num4 in zip(lst_3, lst_4):
    print(f'num3: {num3}, num4: {num4}')
```

```python
for num3, num4 in zip_longest(lst_3, lst_4, fillvalue='not found'):
    print(f'num3: {num3}, num4: {num4}')
```
